#### Background

 

As shown in the review of World Cancer Research International Dataset, the most frequent malignancy in women worldwide is breast cancer. Why this is the case is unknown. Just like most cancers, breast cancer is a heterogeneous disease meaning that there are many factors that can contribute to its onset of breast cancer [1.1] Environmental factors (diet, pollutants, exposure) as well as molecular features (hormones, growth factors, genetic BRCA mutations) have been found to increase the likelihood of disease development. #### Correlation between cancer features 

It is interesting to note the relationships that arise from the correlation table below.  

Similar trends are seen between malignant tumors and benign tumors. Not surprisingly, there is an obvious relationship between the area of the tumor, the radius, and the perimeter. As proven by many mathematicians there is a direct relationship between the radius and the perimeter of many shapes. Archimedes, accredited with the algorithm for calculating the perimeter of a circle 2πr and the area A=πr2 (r = radius) for example [1.9 -1.10].  

 

From the tables below there does not seem to be an obvious relationship between the shape (area, radius, perimeter) of the tumor with any of the tumor distinctions such as smoothness, symmetry, compactness, or fractal dimension. There is a slight increase in the concavity and concavity points seen in malignant tumor where a correlation greater than or = 0.49 is apparent. This is interesting as the relationship between shape and concavity is not as apparent in benign tumors. 

```python
plt.figure(figsize=(30,14)) 

sns.heatmap(malignant_sub.corr(),annot=True,cmap='Reds', annot_kws={'size': 20})
```

```python
plt.figure(figsize=(30,14)) 

sns.heatmap(benign_sub.corr(),annot=True,cmap='Purples', annot_kws={'size': 20}) 
```

#### 2.4. Using Graphs to show relationships 

The scatterplots below indicate that malignant tumors, when compared with benign tumors, have a greater area, radius, and perimeter. The smoothness and symmetry are quite evenly matched between the both cancer types however the compactness and concavity appear to increase with malignancy. 

```python
def area_graphs (yaxis, position): 

    plt.subplot(2, 3, position) 

    sns.scatterplot(data = malignant_sub, x = "area_mean", y = yaxis, legend = True, color = 'red') 

    sns.scatterplot(data = benign_sub, x= "area_mean", y = yaxis, legend = True, color = 'purple') 

    plt.legend(labels=["Malignant", "Benign "]) 

     

fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(10, 7), tight_layout=False) 

area_graphs("perimeter_mean", 1) 

plt.title("Area vs Perimeter") 

  

area_graphs("radius_mean", 2) 

plt.title("Area vs Radius") 

  

area_graphs("smoothness_mean", 3) 

plt.title("Area vs Smoothness") 

  

area_graphs("symmetry_mean", 4) 

plt.title("Area vs Symmetry") 

  

area_graphs("compactness_mean", 5) 

plt.title("Area vs Compactness") 

  

area_graphs("concavity_mean", 6) 

plt.title("Area vs Concavity") 

  

  

plt.show() 

plt.clf() 
```

#### 2.4 Histograms - distribution relationship 

The Oxford dictionary defines distribution as "the action of sharing something out among a number of recipients."[1.11] Histograms can be used to show the distribution of a particular value amongst certain parameters. For example, the first histogram below shows that of all the cases within the dataset malignant_sub most of the population greater than 500 but less than 1500.   

  

   

  

The first observation is that the benign tumor distributions seem to form more of bell-shaped distribution than the malignant dataset. Though this evaluation is open to bias, it appears that the following distribution is evident amongst the dataset [1.12 - 1.14]:  

  

   

  

Random Distribution: Malignant (6)- area, perimeter, radius, concavity, smoothness, symmetry.   

  

Bimodal Distribution: Malignant (1) - compactness.   

  

Bell-shaped: Benign (5) - area, perimeter, radius, smoothness, symmetry.   

  

Skewed-right (positively skewed): Benign (2) - compactness, concavity. 



```python
fig, axes = plt.subplots(nrows=7, ncols=2, figsize=(15, 15), tight_layout=False) 

  

def distributionGraphs_malignant(measurement, position): 

    plt.subplot(7, 2, position) 

    plt.hist( malignant_sub[measurement], color = 'red', edgecolor = 'black',bins = int(180/5)) 

    plt.xlabel(measurement, size = 9) 

    plt.ylabel('Number of cases', size = 9) 

     

def distributionGraphs_benign(measurement, position): 

    plt.subplot(7, 2, position) 

    plt.hist( benign_sub [measurement], color = 'purple', edgecolor = 'black',bins = int(180/5)) 

    plt.xlabel(measurement, size = 9) 

    plt.ylabel('Number of cases', size = 9) 

  

distributionGraphs_malignant('area_mean', 1) 

plt.title('Hist: mean area of Malignant Case', size = 9) 

distributionGraphs_malignant('perimeter_mean', 3) 

plt.title('Hist: mean perimeter of Malignant Case', size = 9) 

distributionGraphs_malignant('radius_mean', 5) 

plt.title('Hist: mean radius of Malignant Case', size = 9) 

  

distributionGraphs_malignant('concavity_mean', 7) 

plt.title('Hist: mean concavity of Malignant Case', size = 9) 

distributionGraphs_malignant('smoothness_mean', 9) 

plt.title('Hist: mean smoothness of Malignant Case', size = 9) 

distributionGraphs_malignant('symmetry_mean', 11) 

plt.title('Hist: mean symmetry of Malignant Case', size = 9) 

distributionGraphs_malignant('compactness_mean', 13) 

plt.title('Hist: mean compactness of Malignant Case', size = 9) 

  

distributionGraphs_benign('area_mean', 2) 

plt.title('Hist: mean area of Benign Case', size = 9) 

distributionGraphs_benign('perimeter_mean', 4) 

plt.title('Hist: mean perimeter of Benign Case', size = 9) 

distributionGraphs_benign('radius_mean', 6) 

plt.title('Hist: mean radius of Benign Case', size = 9) 

  

distributionGraphs_benign('concavity_mean', 8) 

plt.title('Hist: mean concavity of Benign Case', size = 9) 

distributionGraphs_benign('smoothness_mean', 10) 

plt.title('Hist: mean smoothness of Benign Case', size = 9) 

distributionGraphs_benign('symmetry_mean', 12) 

plt.title('Hist: mean symmetry of Benign Case', size = 9) 

distributionGraphs_benign('compactness_mean', 14) 

plt.title('Hist: mean compactness of Benign Case', size = 9) 

 
```
